```python
import numpy as np
import matplotlib.pyplot as plot

from pyclouds.common import default_constants

%matplotlib inline
```

```python
constants = default_constants

Rv = constants.get('R_v')
Rd = constants.get('R_d')
rho_l = constants.get('rho_l')

def temp(ql, qv):
    p = 101325.
    qd = 1.0 - ql - qv
    inv_T = (1./1.2 - ql/rho_l)*(qd*Rd + qv*Rv)/p
    return 1.0/inv_T
```

```python
x_ = np.linspace(0., 0.01, 100)
qv, ql = np.meshgrid(x_, x_, indexing='ij')
```

```python
plot.pcolormesh(qv, ql, temp(ql, qv) - temp(0., 0.))
plot.ylabel('ql')
plot.xlabel('qv')
plot.colorbar()
```
